# 第二题：使用sklearn的KMeans完成聚类

实验内容：
1. 使用sklearn的Kmeans完成两个数据集的聚类任务
2. 计算外部指标FMI和NMI
3. 对聚类结果可视化

# 1. 导入模块

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

# 2. 生成数据集

```python
from sklearn import datasets
X1, y1 = datasets.make_circles(n_samples = 1500, factor = 0.5, noise = 0.05, random_state = 32)
X2, y2 = datasets.make_moons(n_samples = 1500, noise = 0.05, random_state = 32)
```

```python
colors = np.array(['#377eb8', '#ff7f00', '#4daf4a', '#f781bf', '#a65628', '#984ea3', '#999999', '#e41a1c', '#dede00'])
```

```python
plt.title('dataset1')
plt.scatter(X1[:, 0], X1[:, 1], s = 10, c = colors[y1])
```

```python
plt.title('dataset2')
plt.scatter(X2[:, 0], X2[:, 1], s = 10, c = colors[y2])
```

# 3. 导入模型

我们导入KMeans模型

```python
from sklearn.cluster import KMeans
```

# 4. 聚类

```python
model = KMeans(n_clusters=2)
model.fit(X1)
```

使用predict方法预测结果

```python
model.predict(X1)
```

# 5. 聚类结果可视化

```python
plt.title('dataset1')
plt.scatter(X1[:, 0], X1[:, 1], s = 10, color=colors[model.predict(X1)])
```

```python
# YOUR CODE HERE
from sklearn.metrics import normalized_mutual_info_score
from sklearn.metrics import fowlkes_mallows_score
```

```python
# 数据集1
model = KMeans(n_clusters=2)
model.fit(X1)
model.predict(X1)

# 可视化
plt.title('dataset1')
plt.scatter(X1[:, 0], X1[:, 1], s = 10, color=colors[model.predict(X1)])

# NMI
n1 = normalized_mutual_info_score(y1, model.predict(X1))

# FMI
f1 = fowlkes_mallows_score(y1, model.predict(X1))

print(f1,'|',n1)
```

```python
# 数据集2
model = KMeans(n_clusters=2)
model.fit(X2)
model.predict(X2)

# 可视化
plt.title('dataset2')
plt.scatter(X2[:, 0], X2[:, 1], s = 10, color=colors[model.predict(X2)])

# 数据集2
# NMI
n2 = normalized_mutual_info_score(y2, model.predict(X2))

# FMI
f2 = fowlkes_mallows_score(y2, model.predict(X2))

print(f2, '|', n2)
```
